```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

# What defferince between Barplot and Count plot?

 Barplot caluclate mean of data but countplot work count of data.

# Barplot:-

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
```

# Count plot:-

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=var,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette='bwr')
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color="y")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color="r",
              saturation=0.1)
plt.show()
```
