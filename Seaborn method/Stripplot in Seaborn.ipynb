```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var,color="r")
plt.show()
```

```python
sns.stripplot(x=var.day,y=var.total_bill,hue=var.sex)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",hue="sex",data=var,palette="winter_r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",hue="sex",data=var,
              hue_order=["Female","Male"],palette="rocket_r",linewidth=1)
plt.show()
```

```python
sns.stripplot(y="day",x="total_bill",data=var,hue="sex",
              linewidth=1,edgecolor="red")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
              linewidth=1,edgecolor="red")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
             palette="rocket_r",linewidth=1,edgecolor="r",jitter=1)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",hue="sex",data=var,
              linewidth=1,edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",hue="sex",data=var,
              linewidth=1,edgecolor="r",size=5)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
              linewidth=1,edgecolor="red",size=10)
plt.show()
```

```python
m = {"Male":"*","Female":"o"}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
              palette="rocket_r",linewidth=1,edgecolor="r",
              size=5,marker="<")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
              palette="rocket_r",linewidth=1,edgecolor="r",
              size=5,marker="<",alpha=0.5)
plt.show()
```
