```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y)
plt.savefig("line2",dpi=2000)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line3",dpi=2000,facecolor="g")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line4.jpg",dpi=2000,facecolor="g",transparent=True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line5.jpg",dpi=2000,facecolor="g",transparent=False
            ,bbox_inches="tight")
plt.show()
```
