# Violineplot in Seaborn:-

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,
               split="time",hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"],saturation=0.2)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,color="r")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,
               split=True,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,
               split="time",hue="time")
plt.show()
```

```python
sns.violinplot(y="total_bill",x="day",data=var,
               hue="sex",split=True,scale="width")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,
              hue="sex",split=True,scale="count")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var
               ,hue="sex",split=True,scale="area")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var,
               hue="sex",split=True,scale="width")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"],inner="quart")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"],inner="box")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"],inner="quartile")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=var,order=["Dinner","Lunch"],inner="stick")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill"
               ,data=var,order=["Dinner","Lunch"],inner="point")
plt.show()
```

```python
sns.violinplot(y="total_bill",x="time",
               data=var,order=["Dinner","Lunch"],inner=None)
plt.show()
```

# How to make singel violinplot use this method :-

```python
# x axis along graph then he show horizontal graph because provide singel data.
sns.violinplot(x=var["total_bill"])
plt.show()
```

```python
# y axis along graph then he show verticle graph becauese you provide single data.
sns.violinplot(y=var["total_bill"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",
               data=var,linewidth=2,palette="Dark2")
sns.set(style="darkgrid")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",
               data=var,linewidth=2,palette="Dark2")
sns.set(style="dark")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",
               data=var,linewidth=2,palette="Dark2")
sns.set(style="white")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",
               data=var,linewidth=2,palette="Dark2")
sns.set(style="whitegrid")
plt.show()
```
