```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
sns.boxplot(y=var["total_bill"])
sns.set_style("white")
plt.show()
```

```python
sns.boxplot(y=var["total_bill"])
sns.set_style("dark")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("whitegrid")
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("darkgrid")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("ticks")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("white")
sns.despine()
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("white")
sns.despine()
# plt.figure(figsize=(3,12))
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("darkgrid")
plt.figure(figsize=(4,2))
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var,hue="sex"
            ,hue_order=["Female","Male"])
sns.set_style("darkgrid")
sns.set_context("paper",font_scale=2)
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,palette="cool",kind="bar")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,
            hue="sex",hue_order=["Female","Male"],palette="cool",kind="bar")
sns.set_style("darkgrid")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,
            hue="sex",hue_order=["Female","Male"],palette="cool",kind="bar")
sns.set_style("darkgrid")
plt.figure(figsize=(12,3))
plt.show()
```
