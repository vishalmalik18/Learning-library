# Heatmap in Seaborn:-

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# When you want own data in heatmap use this method:-

```python
var4 = np.linspace(1,10,10).reshape(2,5)
var4
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
               ["b0","b1","b2","b3","b4"]])
```

```python
ar
```

```python
sns.heatmap(var4,vmin=0,vmax=12,cmap='PuOr',annot=ar,fmt="s")
```

```python
z = {"fontsize":20,"color":"r"}
sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=z)
```

```python
a = {"fontsize":20,"color":"r"}
sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',
            annot=True,annot_kws=a,linewidth=4)
```

```python
b = {"fontsize":20,"color":"r"}
sns.heatmap(var4,vmin=0,vmax=10,cmap="PuOr",annot=True
            ,annot_kws=b,linewidth=10,linecolor="y")
```

```python
c = {"fontsize":20,"color":"r"}
sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',
            annot=True,annot_kws=c,linewidth=10,linecolor="y"
            ,cbar=False)
```

```python
d = {"fontsize":20,"color":"r"}
sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',annot=True,
            annot_kws=d,linewidth=10,linecolor="y",
            cbar=False,xticklabels=False,yticklabels = False)
```

```python
e = {"fontsize":20,"color":"r"}
v = sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',
                annot=True,annot_kws=e,linewidth=10,
                linecolor="y",cbar=False,xticklabels=False,
                yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
plt.show()
```

```python
e = {"fontsize":20,"color":"r"}
v = sns.heatmap(var4,vmin=0,vmax=10,cmap='PuOr',
                annot=True,annot_kws=e,linewidth=10,
                linecolor="y",cbar=False,xticklabels=False,
                yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
sns.set(font_scale=1)
plt.show()
```
