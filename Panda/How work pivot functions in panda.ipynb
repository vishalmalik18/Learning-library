# How work pivot functions:-

```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"days":[1,2,3,4,5,6],
                     "st_name":["a","b","c","a","b","c"],
                     "eng":[10,12,13,14,15,16],"maths":[17,18,19,20,21,22]})
```

```python
var1
```

```python
var2 = var1.pivot(index="days",columns="st_name")
```

```python
var2
```

```python
var3 = var1.pivot(index="days",columns="st_name",values="eng")
```

```python
var3
```

# when you want arthmetic opperations use this method:-

```python
var4 = var1.pivot_table(index="st_name",columns="days",aggfunc="mean")
```

```python
var4
```

```python
var5 = var1.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

```python
var5
```

# How wok margin:-

```python
var6 = var1.pivot_table(index="st_name",columns="days",aggfunc="sum",margins="True")
```
