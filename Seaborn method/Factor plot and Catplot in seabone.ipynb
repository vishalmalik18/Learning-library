```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
var2 = sns.load_dataset("tips").head(20)
```

```python
var2
```

```python
sns.pointplot(x="size",y="tip",data=var,hue="sex")
plt.show()
```

```python
sns.pointplot(x="size",y="tip",data=var2,hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var2,hue="sex",kind="point")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var2,
            hue ="sex",palette="Oranges",kind="strip")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var2,
            hue="sex",palette="BuGn",kind="point")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,hue="sex",palette="Oranges"
            ,height=3,kind="point")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,hue=
            "sex",kind="box",height=5,showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"g"})
plt.show()
```

```python
sns.catplot(y="day",x="total_bill",data=var,hue=
            "sex",kind="box",height=5,showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"g"})
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,hue=
            "sex",kind="box",height=10,showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"g"})
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=var,hue=
            "sex",kind="boxen")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue=
            "sex",kind="count")
plt.show()
```
