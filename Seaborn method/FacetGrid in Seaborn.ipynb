```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
var2 = sns.FacetGrid(var,col="sex")
var2.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
var3 = sns.FacetGrid(var,col="sex",hue="sex")
var3.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
var4 = sns.FacetGrid(var,col="day",hue="sex")
var4.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
var5 = sns.FacetGrid(var,col="day",hue="sex",palette="summer")
var5.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
var6 = sns.FacetGrid(var,col="day",hue="sex",palette="summer")
var6.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```
