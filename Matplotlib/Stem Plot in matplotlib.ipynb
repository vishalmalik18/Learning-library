```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+",bottom=2)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+",bottom=1)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+",bottom=0,basefmt="g")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+",
         bottom=0,basefmt="g",use_line_collection=False)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,label="python",linefmt=":",markerfmt="r+",
         bottom=0,basefmt="g",use_line_collection=False,
         orientation="horizontal")
plt.legend()
plt.show()
```
