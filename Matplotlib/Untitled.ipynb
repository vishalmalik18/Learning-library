```python
import matplotlib.pyplot as plt
```

```python
x = ["a","s","d","f"]
y = [10,20,30,40]
plt.subplot(2,2,1)
plt.bar(x,y)

x1 = [10,20,30,40]
y1 = ["python","java","c++","c+"]
z1 = [0.1,0.0,0.0,0]
plt.subplot(2,2,3)
plt.pie(x1,labels=y1,explode=z1)

x1 = [1,2,3,4]
y1 = [1,2,3,4]
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.show()
```
