```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]
plt.step(x,y,color="r",marker="o",mfc="g",ms=10,label="python",linestyle="-"
         ,alpha=0.5)
plt.legend(loc=2)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
