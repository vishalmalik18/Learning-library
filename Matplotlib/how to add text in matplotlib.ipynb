```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
z = [1,2,3,4,6]
j = [2,3,5,6,2]
plt.plot(x,y)
plt.plot(z,j)
plt.plot()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["python","java"])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("coding_comunity",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("python",fontsize=15)
plt.text(2,5,"python",fontsize=15,style="italic",bbox={"facecolor":"red"})
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("coding_comumniy")
plt.xlabel("days")
plt.ylabel("python")
plt.annotate("python",xy=(2,1),xytext=(3,3),
             arrowprops=dict(facecolor="black",shrink=100))
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("coding_community")
plt.xlabel("days")
plt.ylabel("python")
plt.legend(["python"],facecolor="red",edgecolor="yellow",framealpha=0.5,shadow=True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("coding_community")
plt.xlabel("days")
plt.ylabel("python")
plt.legend(["python"],loc=1)
plt.show()
```
