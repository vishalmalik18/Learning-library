```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,4,5,2,4]
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,4,5,2,4]
plt.plot(x,y)
plt.xticks(x,labels=["python","java","c++","c","ruby"])
plt.yticks(y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,4,5,2,4]
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,4,5,2,4]
plt.plot(x,y)
plt.axis([0,10,0,7])
plt.show()
```
