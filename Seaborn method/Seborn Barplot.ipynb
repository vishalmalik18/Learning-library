```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style="darkgrid")
```

```python
var = sns.load_dataset("penguins")
```

```python
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data= var)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,order= order_1)
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,order= order_1,hue_order=["Female","Male"])
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,
            order=order_1,hue_order=["Female","Male"],ci=90)
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",ci=100,n_boot=100)
plt.show()
```

# Vertical graph:-

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm"
            ,data=var,orient="v")
plt.show()
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm"
            ,data=var,orient="h")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,color="g")
plt.show()
```

# When you change colour bar use this method :-

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",order=order_1,
            hue_order=["Female","Male"],orient="v"
            ,palette="icefire")
plt.show()
```

# When you want blur use this method:-

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",saturation=0.3)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",saturation=0.5)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",saturation=10)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",errcolor="y")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data= var,
            hue="sex",order=order_1,hue_order=["Female","Male"],
            orient="v",saturation=10,errcolor="y")
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,
            hue="sex",orient="v",saturation=10,errcolor="y",errwidth=10)
plt.show()
```

# Use line show end point use this method:-

```python
sns.barplot(x="island",y="bill_length_mm"
            ,data=var,hue="sex",orient="v",saturation=10,capsize=0.5)
plt.show()
```

# Two graph superinpose:-

```python
sns.barplot(x="island",y="bill_length_mm",data=var,
            hue="sex",orient="v",saturation=10,dodge=False)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",
            saturation=10,errcolor="b",)
sns.set(style="darkgrid")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",
            data=var,hue="sex",orient="v",
            saturation=10,errcolor="b",alpha=0.5)
sns.set(style="darkgrid")
plt.show()
```
