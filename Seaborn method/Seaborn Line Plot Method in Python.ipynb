```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var1 = [1,2,3,4,5,6,7]
var2 = [2.5,3.5,4.0,2.0,4.0,5.0,6.0]
var3 = pd.DataFrame({"var1":var1,"var2":var2})
```

```python
var3
```

# Seaborn method:-

```python
sns.lineplot(x="var1",y="var2",data=var3)
plt.show()
```

# Matploatlib method:-

```python
plt.plot(var1,var2)
plt.show()
```

```python
var4 = sns.load_dataset("penguins")
```

```python
var4
```

# Liner plot

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data= var4,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=var4,hue="sex",palette='Accent')
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data = var4,palette= 'Accent' ,hue = "sex" ,markers= ["o",">"])
plt.show()
```
