# How work melt functions:-

```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"Days":[1,2,3,4,5,6],"maths":[55,76,78,90,40,33],"english":[22,33,44,55,66,77]})
```

```python
var1
```

```python
var2 = pd.melt(var1)
```

```python
var2
```

```python
var3 = pd.melt(var1,id_vars=["Days"])
```

```python
var3
```

```python
var4 = pd.melt(var1,id_vars=["Days"],var_name="python")
```

```python
var4
```

```python
var5 = pd.melt(var1,id_vars=["Days"],var_name="python",value_name="wscube")
```

```python
var5
```
