```python
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,4,3]
plt.plot(x,color="r",marker="o",mfc="g",ms=10,label="woman")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
x = [2,4,6,4,3]
plt.plot(x,color="r",marker="o",mfc="g",ms=10,linestyle="--",label="Woman")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
x = [2,4,6,4,3]
y = [3,7,8,3,4]
plt.plot(x,linestyle="--",marker="o",color="r",label="woman")
plt.plot(y,color="b",marker="+",label="man")
plt.legend()
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.legend()
plt.show()
```

```python
x = [2,4,6,4,3]
y = [3,7,8,3,4]
plt.plot(x,linestyle="--",marker="o",color="r",label="woman",alpha=0.7)
plt.plot(y,color="b",marker="+",label="man",alpha=0.8)
plt.legend()
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
