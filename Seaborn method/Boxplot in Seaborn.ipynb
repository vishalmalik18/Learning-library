```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=var,color="g")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",
            order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",
            color="g",showmeans=True)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,color="g")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans = True,
            meanprops={"marker":"+","markeredgecolor":"b"},palette="BuGn")
plt.show()
       
```

```python
sns.boxplot(x="day",y="total_bill",data=var,hue = "sex",
            hue_order=["Female","Male"],showmeans = True,
            meanprops={"marker":"+","markeredgecolor":"b"},
            palette="plasma")
plt.show()
```

# Orient parmater:-

# x axis along:-

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},orient="v",
            color="g")
plt.show()
```

# Without orient:-

# x axis along:-

```python
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},color="g")
plt.show()
```

# y axis along:-

```python
sns.boxplot(y="day",x="total_bill",data=var,hue="sex",showmeans=True
            ,meanprops={"marker":"+","markeredgecolor":"g"},color="g")
plt.show()
```
