```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.title("python")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.plot(x,y)
plt.fill_between(x,y,label="python",color="g",alpha=0.5)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.title("python")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4,label="python")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.title("python")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.plot(x,y)
plt.fill_between(x,y,label="python",color="g",where=(x>=2)&(x<=4))
plt.legend()
plt.show()
```
