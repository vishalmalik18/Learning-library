```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
area1 = [2,3,4,5,6]
plt.stackplot(x,area1)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
area1 = [2,4,5,7,3,3,2,4]
area2 = [5,7,6,4,3,4,1,5]
area3 = [2,6,4,3,1,2,7,3]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.title("python")
plt.show()
```
