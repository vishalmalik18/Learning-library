# When you want multiple bar graph use this function

```python
x = ["python","c","c++","java"]
y = [80,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("langugae",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Data",fontsize=15)

plt.bar(p,y,width,color="r",label="pouplarity")
plt.bar(p1,z,width,color="y",label="pouplarity1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

# When you change rotation use this method :-

```python
x = ["python","c","c++","java"]
y = [80,70,60,82]
z = [20,30,40,50]

width = 0.2

p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Data",fontsize=15)

plt.bar(p,y,width,color="r",label="pouplarity")
plt.bar(p1,z,width,color="y",label="pouplariy1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
