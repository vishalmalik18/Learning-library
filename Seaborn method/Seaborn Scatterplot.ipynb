# Seborn Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
```

```python
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,hue=var.sex)
plt.show()
```

```python
var2 = sns.load_dataset("penguins").head(20)
```

```python
var2
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
                data=var,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
                data=var2,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
                data = var2,hue="sex",style="sex",size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
                data=var2,hue="sex",style="sex",
                size="sex",sizes=(100,40))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
                data=var,hue="sex",style="sex",
                size="sex",sizes=(100,40))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var,hue="sex",style="sex",
               size="sex",sizes=(100,40),palette="gist_ncar")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var2,hue="sex",style="sex",
               size="sex",sizes=(100,40),palette="gist_ncar")
plt.show()
```

# With style:-

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var,hue="sex",style="sex",
               size="sex",sizes=(100,40),palette="gist_ncar")
plt.show()
```

# Without style:-

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var,hue="sex",
               size="sex",sizes=(100,40),palette="gist_ncar")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data= var,
                hue="sex",style="sex",size="sex",sizes=(100,40),
                palette="gist_ncar",alpha=0.5)
plt.show()
```

```python
n = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var,hue="sex",
               size="sex",sizes=(100,120),style="sex",palette="gist_ncar",markers=n)
plt.show()
```

```python
n = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",
               data =var,hue="sex",
               size="sex",sizes=(100,60),style="sex",palette="gist_ncar",markers=n)
plt.show()
```
