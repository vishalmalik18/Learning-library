```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
```

```python
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",diag_kind="hist")
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",kind="kde")
```

```python
sns.pairplot(var,vars=["total_bill","tip"],
             hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",
             hue_order=["Female","Male"],palette="BuGn")
plt.show()
```

# x axis along:-

```python
sns.pairplot(var,vars=["total_bill","tip"],
             hue="sex",hue_order=["Female","Male"],
             x_vars=["total_bill","tip"])
plt.show()
```

# y axis along:-

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"]
             ,y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="hist")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="scatter")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="kde")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],
             kind="kde",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="scatter",diag_kind="kde")
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],
             kind="scatter",diag_kind="hist")
plt.show()
```
