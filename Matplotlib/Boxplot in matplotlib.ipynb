```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
x1 = [10,20,30,40,50,60,70]
y = [x,x1]
plt.boxplot(y,labels=["python","c++"])
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,vert=False,labels=["python"])
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],widths=0.5)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],notch=True)
plt.show()
```

```python
x = [10,20,30,40,50,60]
plt.boxplot(x,labels=["python"],patch_artist=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],showmeans=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+")
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,sym=" ")
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",boxprops=dict(color="r"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="g"),
            whiskerprops=dict(color="r"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="g"),
            whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="g"),
            whiskerprops=dict(color="r"))
plt.show()
```
